```python
import cv2
import matplotlib.pyplot as plt
import numpy as np


img_path = "Covid19Pos-LungCT.png"
img = cv2.imread(img_path)
```

```python
def low_pass_filter(img_path, kernel_size):
    img = cv2.imread(img_path)
    kernel = np.ones([kernel_size, kernel_size], dtype=int)
    kernel_sum = sum(kernel)
    kernel = kernel / kernel_sum
    img_rst = cv2.filter2D(img, -1, kernel)
    plt.subplot(1,2,1), plt.imshow(img, cmap="gray")
    plt.title("Original"), plt.xticks([]), plt.yticks([])
    plt.subplot(1,2,2), plt.imshow(img_rst, cmap="gray")
    plt.title("Low pass Filter"), plt.xticks([]), plt.yticks([])
```

```python
low_pass_filter(img_path, 1)
```

```python
def averaging(img_path, kernel_size):
    img = cv2.imread(img_path)
    blur = cv2.blur(img,(kernel_size, kernel_size))
    plt.subplot(1,2,1), plt.imshow(img, cmap="gray")
    plt.title("Original"), plt.xticks([]), plt.yticks([])
    plt.subplot(1,2,2), plt.imshow(blur, cmap="gray")
    plt.title("Average Filter"), plt.xticks([]), plt.yticks([])   
```

```python
averaging(img_path, 10)
```

```python
def gaussian(img_path, kernelx, kernely):
    img = cv2.imread(img_path)
    blur = cv2.GaussianBlur(img, (kernelx, kernely), 0)
    plt.subplot(1,2,1), plt.imshow(img, cmap="gray")
    plt.title("Original"), plt.xticks([]), plt.yticks([])
    plt.subplot(1,2,2), plt.imshow(blur, cmap="gray")
    plt.title("Gaussian Filter output"), plt.xticks([]), plt.yticks([])
    
```

```python
gaussian(img_path, 3, 3)
```

```python
# cv2.medianBlur() computes the median of all the pixels under the kernel
## window and the central pixel is replaced with this median value
def median(img_path, kernel):
# The kernel size must be a positive odd integer.
    img = cv2.imread(img_path)
    blur = cv2.medianBlur(img, kernel)
    plt.subplot(1,2,1), plt.imshow(img, cmap="gray")
    plt.title("Original"), plt.xticks([]), plt.yticks([])
    plt.subplot(1,2,2), plt.imshow(blur, cmap="gray")
    plt.title("Median Filter output"), plt.xticks([]), plt.yticks([])
```

```python
median(img_path, 3)
```

```python
def bilateralFilter (img_path):
    img = cv2.imread(img_path)
    blur = cv2.bilateralFilter(img,9,75,75)
    plt.subplot(1,2,1), plt.imshow(img, cmap="gray")
    plt.title("Original"), plt.xticks([]), plt.yticks([])
    plt.subplot(1,2,2), plt.imshow(blur, cmap="gray")
    plt.title("Bilateral Filter output"), plt.xticks([]), plt.yticks([])
    
```

```python
bilateralFilter(img_path)
```

```python
def gradient(img_path):
    img = cv2.imread(img_path,0)

    laplacian = cv2.Laplacian(img,cv2.CV_64F)
    sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
    sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)

    plt.subplot(2,2,1),plt.imshow(img,cmap = 'gray')
    plt.title('Original'), plt.xticks([]), plt.yticks([])
    plt.subplot(2,2,2),plt.imshow(laplacian,cmap = 'gray')
    plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
    plt.subplot(2,2,3),plt.imshow(sobelx,cmap = 'gray')
    plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
    plt.subplot(2,2,4),plt.imshow(sobely,cmap = 'gray')
    plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])

    plt.show()
```

```python
gradient(img_path)
```

```python
def canny(img_path):
    img = cv2.imread(img_path,0)
    edges = cv2.Canny(img,100,200)

    plt.subplot(121),plt.imshow(img,cmap = 'gray')
    plt.title('Original Image'), plt.xticks([]), plt.yticks([])
    plt.subplot(122),plt.imshow(edges,cmap = 'gray')
    plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

    plt.show()
```

```python
canny(img_path)
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np


img_path = "Covid19Pos-LungCT.png"
img = cv2.imread(img_path)
plt.imshow(img)
plt.show()
```

```python
def low_pass_filter(img_path, kernel_size):
    img = cv2.imread(img_path)
    kernel = np.ones([kernel_size, kernel_size], dtype=int)
    kernel_sum = sum(kernel)
    kernel = kernel / kernel_sum
    img_rst = cv2.filter2D(img, -1, kernel)
    plt.imshow(img_rst)
    plt.show()
#     return img_rst
```

```python
low_pass_filter(img_path, 1)
```

```python
def averaging(img_path, kernel_size):
    img = cv2.imread(img_path)
    blur = cv2.blur(img,(kernel_size, kernel_size))
    plt.imshow(blur)
    plt.show()
#     return blur    
```

```python
averaging(img_path, 1)
```

```python
def gaussian(img_path, kernelx, kernely):
    img = cv2.imread(img_path)
    blur = cv2.GaussianBlur(img, (kernelx, kernely), 0)
    plt.imshow(blur)
    plt.show()
    
```

```python
gaussian(img_path, 3, 3)
```

```python
# cv2.medianBlur() computes the median of all the pixels under the kernel
## window and the central pixel is replaced with this median value
def median(img_path, kernel):
#     The kernel size must be a positive odd integer.
    img = cv2.imread(img_path)
    blur = cv2.medianBlur(img, kernel)
    plt.imshow(blur)
    plt.show()
```

```python
median(img_path, 3)
```

```python
def bilateralFilter (img_path):
    img = cv2.imread(img_path)
    blur = cv2.bilateralFilter(img,9,75,75)
    plt.imshow(blur)
    plt.show()
```

```python
bilateralFilter(img_path)
```

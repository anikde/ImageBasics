```python
import matplotlib.pyplot as plt

from skimage.feature import graycomatrix, graycoprops
from skimage import io
import cv2
import numpy as np
```

```python
imagePath = "Covid (20).png"
image = cv2.imread(imagePath, cv2.IMREAD_GRAYSCALE)
plt.imshow(image)
plt.show()
print(f"shape : {image.shape}")
print(f"type : {type(image)}")
print(image.dtype)
```

```python
image = cv2.resize(image, (248, 248))
print(image.shape)
```

```python
subimage = image[0:150, 0:100]
plt.imshow(subimage)
plt.show()
```

```python
ksize = 20
sigma = 10
lamda = 1 * np.pi/4
theta = 1 * np.pi/4
gamma = 0.5
phi = 0
```

```python
kernel = cv2.getGaborKernel((ksize, ksize), sigma, theta, lamda, gamma, phi, ktype=cv2.CV_32F)
plt.imshow(kernel)
plt.show()
```

```python
fimage = cv2.filter2D(subimage, cv2.CV_8UC3, kernel)
plt.imshow(fimage)
plt.show()
```
